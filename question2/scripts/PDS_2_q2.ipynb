```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a visual theme
sns.set_theme(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

df.to_csv('student_performance_processed.csv', index=False)
# Display first few rows
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Visualization 1: Histogram of Math Scores
plt.figure(figsize=(8, 6))
plt.hist(df["math score"], bins=15, color="skyblue", edgecolor="black")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
plt.savefig("../content/figures/viz1_histogram.png")
plt.close()
print("Histogram saved.")

# Analysis: Shows the spread and frequency of math scores, helping identify common performance ranges or skewness.
```

```python
# Visualization 2: Scatter Plot of Reading Score vs Writing Score
plt.figure(figsize=(8, 6))
plt.scatter(df["reading score"], df["writing score"], alpha=0.5, color="green")
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
plt.savefig("../content/figures/viz2_scatter.png")
plt.close()
print("Scatter plot saved.")
# Analysis: Reveals the relationship between reading and writing skills, highlighting correlations or outliers.
```

```python
# Visualization 3: Boxplot of Math Score by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x="gender", y="math score", data=df, palette="Set2")
plt.title("Math Score Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
plt.savefig("../content/figures/viz3_boxplot.png")
plt.close()
print("Boxplot saved.")
# Analysis: Compares math performance between genders, showing median, spread, and potential disparities.
```

```python
# Visualization 4: Bar Plot of Average Writing Score by Parental Education
plt.figure(figsize=(10, 6))
sns.barplot(x="parental level of education", y="writing score", data=df, ci=None, palette="Blues")
plt.title("Average Writing Score by Parental Level of Education")
plt.xlabel("Parental Education")
plt.ylabel("Average Writing Score")
plt.xticks(rotation=45)
plt.show()
plt.savefig("../content/figures/viz4_barplot.png")
plt.close()
print("Bar plot saved.")
# Analysis: Assesses how parental education impacts average writing performance, identifying trends across education levels.
```

```python
# Visualization 5: Heatmap of Correlations Between Scores
plt.figure(figsize=(8, 6))
correlation_matrix = df[["math score", "reading score", "writing score"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Test Scores")
plt.show()
plt.savefig("../content/figures/viz5_heatmap.png")
plt.close()
print("Heatmap saved.")
# Analysis: Quantifies relationships between math, reading, and writing scores, aiding in understanding subject interdependencies.
```
